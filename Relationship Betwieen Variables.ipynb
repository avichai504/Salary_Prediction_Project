```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("clean data/web developer clean data 3000.csv")
df_copy = df.copy()
df_copy = df_copy[df_copy['Education'] != -1]
```

```python

plt.scatter(df['Experience'], df['Annual Salary'])
plt.xlabel('Experience')
plt.ylabel('Annual Salary')
plt.show()
```

```python

# Define markers for each level of education
markers = {0: 'o', 1: 's', 2: '^', 3: 'd'}

# Map education levels to categories
df_copy['Education Category'] = df_copy['Education'].map({0: 'High School', 1: 'Bachelor\'s Degree', 2: 'Master\'s Degree', 3: 'PhD'})

# Create scatter plot with markers based on education level
plt.scatter(df_copy['Experience'], df_copy['Annual Salary'],
            marker=df_copy['Education'].apply(lambda x: markers[x]), s=100)
plt.xlabel('Experience')
plt.ylabel('Annual Salary')
plt.title('Experience vs Annual Salary by Education Level')
plt.legend(tuple(df['Education Category'].unique()), loc='upper left')
plt.show()
```
