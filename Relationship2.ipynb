```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("marge software engineer.csv")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df[df['Experience'] < 12]
df.drop_duplicates(keep='first', inplace=True)
df.shape
```

```python
sns.violinplot(x=df['Education'], y=df['Annual Salary'])
```

```python
sns.violinplot(x=df['Experience'], y=df['Annual Salary'])
```
